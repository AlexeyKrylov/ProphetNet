```python
import os
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
with open("./train-1863570.log") as f:
    data = f.readlines()
    
split_indexes = []
for i in range(len(data)):
    if '-----------------------------\n' in data[i] or '----------------------------------\n' in data[i]:
        split_indexes.append(i)
        
logs = []
for i in range(1, len(split_indexes)):
    logs.append(data[split_indexes[i-1]:split_indexes[i]])
logs = [dict([j.strip().replace(" ", '')[1:-1].split('|') for j in i[1:]]) for i in logs if len(i) > 5]
logs = pd.DataFrame(logs)
logs = logs[~logs['step'].isna()]
logs = logs.astype('float64')
```

```python
for col in logs.columns:
    sns.lineplot(data=logs, x="step", y=col)
    plt.show()
```
